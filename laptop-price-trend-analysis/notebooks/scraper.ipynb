```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
for i in range(1, 21):
    data = requests.get(f"https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page={i}")
    with open(f"../data/html/page{i}.html", "w", encoding="utf-8") as f:
        f.write(data.text)
        print(f"Downloaded page{i} successfully.")
```

```python
laptops = []

for i in range(1, 21):
    with open(f"../data/html/page{i}.html", "r", encoding="utf-8") as f:
        content = f.read()
     
    soup = BeautifulSoup(content,"html.parser")
    data = soup.select("div.product-wrapper")
     
    for item in data:
        title = item.find('a')['title']
        description = item.find('p', class_='description').text
        price = item.find('span').text.split('$')[1]
        review = item.select('p[data-rating]')[0]['data-rating']
        review_count = item.find(class_='ratings').find('p').text.split()[0]
        
        laptops.append([title, description, price, review, review_count])
```

```python
df = pd.DataFrame(laptops, columns=['Title', 'Description', 'Price', 'Rating', 'Review Count'])
df
```

```python
df.to_csv("../data/cleaned_laptops_data.csv", index=False)
```
