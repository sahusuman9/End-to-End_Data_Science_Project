```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_new = pd.read_csv("../data/cleaned_laptops_data.csv")
df = df_new.copy()
df
```

```python
df.info()
```

```python
df.describe()
```

## DATA TRANSFORMATION

```python
#peice category
df['Price_Category'] = np.where(
    df['Price'] < 500, 'Low',
    np.where(df['Price'] < 1000, 'Medium', 'High')
)
```

```python
df
```

```python
#rating_category
df['Rating_Category'] = np.where(
    df['Rating'] <= 2, 'Low',
    np.where(df['Rating'] <= 3, 'Medium', 'High')
)
```

```python
df
```

## DATA VISUALIZATION AND INSIGHTS

```python
#Histogram: Price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True)
plt.title('Price Distribution of Laptops')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

### Bimodal distribution with peaks at:
    1. $400â€“$500 (â‰ˆ12 laptops)
    2. $1100â€“$1200 (â‰ˆ18 laptops)

### Price range spans from $300 to $1800, indicating wide product variety.

### Mid-range segment ($600â€“$900) is sparse, with only ~5â€“7 laptops.

### High-end outliers above $1500 are rare (â‰ˆ2â€“3 laptops).

### Most laptops fall in the $400â€“$1300 range, covering both budget and premium tiers.

### Right-skewed distribution due to few expensive products.

```python
#Bar chart: Average price per rating
plt.figure(figsize=(10, 6))
sns.barplot(x='Rating_Category', y='Price', data=df, errorbar=None)
plt.title('Average Price per Rating Category')
plt.xlabel('Rating Category')
plt.ylabel('Average Price')
```

### Medium-rated laptops have the highest average price
    Average: â‰ˆ â‚¹950â€“â‚¹980

### Low-rated laptops are moderately priced
    Average: â‰ˆ â‚¹920â€“â‚¹940

### Highly rated laptops have the lowest average price
    Average: â‰ˆ â‚¹820â€“â‚¹840

```python
#Pie chart: Product count in each price category
plt.figure(figsize=(8, 8))
price_counts = df['Price_Category'].value_counts()
plt.pie(price_counts, labels=price_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Product Count in Each Price Category')
```

### High-priced laptops dominate the catalog
    1. 55.6% of all products fall under the High price category (â‚¹1000 and above)
    2. This suggests the store has a premium product focus

### Low-priced laptops make up ~1/3 of inventory
    1. 32.5% are categorized as Low (below â‚¹500)
    2. Indicates a decent offering for budget-conscious customers

### Mid-range laptops are least available
    1. Only 12.0% fall into the Medium category (â‚¹500â€“â‚¹999)
    2. Points to a possible gap in mid-tier offerings, which could be explored further

```python
#Heatmap: Rating vs review count
plt.figure(figsize=(10, 6))
sns.heatmap(df.pivot_table(index='Rating_Category', columns='Price_Category', values='Review Count', aggfunc='sum'), annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Review Count by Rating and Price Category')
```

### Most reviews (236): High-priced, Low-rated laptops â€” suggests customer dissatisfaction drives feedback.

### Next highest (162): High-priced, Medium-rated â€” indicates frequent interaction with premium models.

### Highly-rated laptops receive fewer reviews overall â€” possibly due to customer satisfaction or fewer units.

### Least reviews (8): Medium-priced, Highly-rated laptops â€” likely due to low inventory in this segment.

### Low-priced laptops have balanced review activity (~77â€“86) across all rating tiers.
